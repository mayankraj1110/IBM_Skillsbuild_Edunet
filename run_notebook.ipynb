### 3. Expose the App via a Public Link
Streamlit runs on port `8501`. We will run the Streamlit server in the background and use `localtunnel` via `npx` to generate a direct, secure public link.

```python
import subprocess
import time
import urllib.request

# Start Streamlit app in the background
print("Starting Streamlit app on port 8501...")
streamlit_process = subprocess.Popen([
    "streamlit", "run", "app.py", 
    "--server.port=8501", 
    "--server.address=localhost", 
    "--server.headless=true"
])
time.sleep(5)  # Allow Streamlit server to boot up

# Get Colab public IP (required by localtunnel as a password)
try:
    ip = urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip()
    print(f"Your Tunnel Password (IP Address): {ip}")
except Exception as e:
    print("Could not retrieve public IP:", e)

print("\nStarting public tunnel... Click the URL generated below, enter the IP address above if prompted, and click Submit.")
# Start localtunnel tunnel
!npx localtunnel --port 8501
```
